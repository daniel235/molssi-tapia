```python
import os
import numpy
import matplotlib.pyplot as plt
```

```python
distance_files = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_files, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
print(data)
```

```python

for i in range(1,len(headers)):
    plt.figure()
    plt.plot(data[:,i], label=headers[i])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance(angstrom)')
    plt.legend()
    plt.savefig(F'{headers[i]}.png')
    

```
